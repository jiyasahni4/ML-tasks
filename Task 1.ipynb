```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gaussian_kde
```

```python
# Fréchet distribution parameters
alpha = 3.0  # Shape parameter
gamma = 1.0  # Scale parameter
```

```python
# Number of samples
num_samples = 1000
```

```python
# Generate samples from the Fréchet distribution using Inverse CDF sampling
y = np.random.uniform(size=num_samples)

# Inverse cdf function for frechet
samples = gamma * (-np.log(y)) ** (-1/alpha)
```

```python
# Generate x values for PDF and KDE estimation
x_val = np.linspace(np.min(samples), np.max(samples), num=500)
```

```python
# Compute PDF values for Fréchet distribution
# Pdf function for frechet
pdf_val = (alpha / gamma) * (x_val / gamma) ** (-alpha - 1) * np.exp(-(x_val / gamma) ** -alpha)
```

```python
# Compute KDE estimation
kde_estimation = gaussian_kde(samples)
kde_val = kde_estimation(x_val)
```

```python
# Plot the KDE and PDF curves
plt.plot(x_val, kde_val, label='KDE',color="blue")
plt.plot(x_val, pdf_val, label='PDF',color="yellow")
plt.xlabel('x')
plt.ylabel('Density')
plt.title('KDE vs. PDF')
plt.legend()
plt.show()
```

```python
from matplotlib.pyplot import *
from numpy import linspace, exp

def calc_frechet_pdf(x,a,s):
    y = (a/s) * ((x/s)**(-1-a)) * exp(- (x/s)**(-a) )
    return y
 
x = linspace(0.00001,4,10**5)
 
fig = figure()
ax = fig.add_subplot(111)
ax.plot(x,calc_frechet_pdf(x,1.,1.),label=r'$\alpha=1;$ $s=1$', linewidth=2)
ax.plot(x,calc_frechet_pdf(x,2.,1.),label=r'$\alpha=2;$ $s=1$', linewidth=2)
ax.plot(x,calc_frechet_pdf(x,3.,1.),label=r'$\alpha=3;$ $s=1$', linewidth=2)
ax.plot(x,calc_frechet_pdf(x,1.,2.),label=r'$\alpha=1;$ $s=2$', linewidth=2)
ax.plot(x,calc_frechet_pdf(x,2.,2.),label=r'$\alpha=2;$ $s=2$', linewidth=2)
ax.plot(x,calc_frechet_pdf(x,3.,2.),label=r'$\alpha=3;$ $s=2$', linewidth=2)
ax.grid(True)
ax.minorticks_on()
ax.legend(loc=1)
ax.set_xlim(0,4)
ax.set_ylim(0,1.5)
ax.set_xlabel("x")
ax.set_ylabel("PDF")
fig.savefig("Frechet_pdf.svg",bbox_inches="tight", pad_inches=.15)
```

2)These distributions are clearly right skewed due to a continuous decrease in its slope with an increase in the value of x  
3) These can be considered as a heavy tailed curve as they decay slowly when the value of x is increased. This means that the probability of  occurence of larger values of x is quite considerable
