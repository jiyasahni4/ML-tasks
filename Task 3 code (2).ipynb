```python
interpolation_columns = ['PM2.5 (ug/m3)', 'PM10 (ug/m3)', 'NO2 (ug/m3)']
interpolation_data = Merged_forecast[['Longitude', 'Latitude'] + interpolation_columns]
```

```python
missing_data = interpolation_data[interpolation_data[interpolation_columns].isnull().any(axis=1)]
non_missing_data = interpolation_data.dropna(subset=interpolation_columns)
```

```python
points = non_missing_data[['Longitude', 'Latitude']].values
values = non_missing_data[interpolation_columns].values
```

```python
from scipy.interpolate import griddata

methods = ['linear', 'nearest', 'cubic']

# Iterate over the methods
for method in methods:
 interpolated_values = griddata(points, values, (missing_data['Longitude'], missing_data['Latitude']), method=method)
```

```python
Merged_forecast.loc[missing_data.index, interpolation_columns] = interpolated_values
Merged_forecast.head()
```

```python
from sklearn.linear_model import LinearRegression

# Create a copy of the dataframe for imputation
imputed_data = Merged_forecast.copy()

# Define the features for regression
features = ['Longitude', 'Latitude']

# Iterate over each target column and perform regression-based imputation
for column in ['PM2.5 (ug/m3)', 'PM10 (ug/m3)', 'NO2 (ug/m3)']:
    # Split data into features (X) and target (y)
    X = imputed_data.dropna(subset=[column])[features]
    y = imputed_data.dropna(subset=[column])[column]

    # Train a regression model
    model = LinearRegression()
    model.fit(X, y)

    # Predict missing values
    X_missing = imputed_data[imputed_data[column].isna()][features]
    y_pred = model.predict(X_missing)

    # Fill in missing values with the predicted values
    imputed_data.loc[imputed_data[column].isna(), column] = y_pred
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Select columns for imputation
imputation_columns = ['PM2.5 (ug/m3)', 'PM10 (ug/m3)', 'NO2 (ug/m3)']

# Create a copy of the dataframe for imputation
imputed_data = Merged_data.copy()

# Iterate over each column for imputation
for column in imputation_columns:
    # Split data into features (X) and target (y)
    X = imputed_data[['Longitude', 'Latitude']]
    y = imputed_data[column]
    
    # Create an imputer to handle missing values
    imputer = SimpleImputer()
    X_imputed = imputer.fit_transform(X)
    
    # Drop rows with missing values in both features and target
    valid_rows = ~np.isnan(y)
    X_imputed_valid = X_imputed[valid_rows]
    y_valid = y[valid_rows]
    
    # Create a regression model and fit it
    regressor = LinearRegression()
    regressor.fit(X_imputed_valid, y_valid)
    
    # Select rows with missing values
    missing_data = imputed_data[imputed_data[column].isnull()]
    
    # Impute missing values by transforming longitude and latitude using the imputer
    missing_data_imputed = imputer.transform(missing_data[['Longitude', 'Latitude']])
    
    # Predict missing values using the regression model
    imputed_values = regressor.predict(missing_data_imputed)
    
    # Update the dataframe with imputed values
    imputed_data.loc[missing_data.index, column] = imputed_values
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
X = Merged_data.drop("station_id", axis=1)  # Drop the "station_id" column
y = Merged_data["station_id"]
```

```python
kf = KFold(n_splits=3, shuffle=True, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import confusion_matrix, mean_squared_error
import numpy as np

tree = DecisionTreeRegressor(random_state=0)

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    tree.fit(X_train, y_train)
    y_pred = tree.predict(X_test)
    
    # Convert predicted values to integers (station_id column)
    y_pred = np.round(y_pred).astype(int)
    
    # Calculate evaluation metrics (confusion matrix and mean squared error)
    confusion = confusion_matrix(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    
    # Print or store the evaluation results
    print("Confusion Matrix:")
    print(confusion)
    print("Mean Squared Error:", mse)
```
